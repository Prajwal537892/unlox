# Unsupervised Learning

```python
# Import numpy for numerical operations and arrays
import numpy as np
# Import matplotlib for visualization
import matplotlib.pyplot as plt
# Import KMeans clustering algorithm from sklearn
from sklearn.cluster import KMeans
```

```python
# Import function to generate artificial clustering dataset
from sklearn.datasets import make_blobs
# Create dataset with 300 samples
# centers=4 means we want 4 natural clusters
# cluster_std controls how spread the clusters are
# random_state ensures reproducibility
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)
```

```python
# Plot the data points
plt.scatter(X[:,0], X[:,1])
# Add title to the plot
plt.title("Raw Data")
# Display the plot
plt.show()
```

```python
# Create KMeans model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
# Train the model on the dataset
kmeans.fit(X)
```

```python
# labels_ gives cluster number assigned to each point
labels = kmeans.labels_
# cluster_centers_ gives coordinates of cluster centroids
centroids = kmeans.cluster_centers_
```

```python
# Plot data points colored by cluster label
plt.scatter(X[:,0], X[:,1], c=labels)
# Plot centroids using red X markers
plt.scatter(centroids[:,0], centroids[:,1], color='red', marker='X', s=200)
# Add plot title
plt.title("K-Means Clustering")
# Show plot
plt.show()
```

```python
# Create empty list to store WCSS values
wcss = []
# Try cluster numbers from 1 to 10
for i in range(1, 11):
 # Create KMeans model with i clusters
 kmeans = KMeans(n_clusters=i, random_state=42)
 # Train model
 kmeans.fit(X)
 # inertia_ returns WCSS value
 wcss.append(kmeans.inertia_)
# Plot number of clusters vs WCSS
plt.plot(range(1,11), wcss)
# Add title
plt.title("Elbow Method")
# Label X-axis
plt.xlabel("Number of Clusters")
# Label Y-axis
plt.ylabel("WCSS")
# Display graph
plt.show()
```
